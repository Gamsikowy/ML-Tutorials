```python
q1 = df['not_distracted'].quantile(.25)
q3 = df['not_distracted'].quantile(.75)
iqr = q3 - q1
h = (2.64 * iqr) / (len(df['not_distracted'])**(1/3))
b = int((df['not_distracted'].max() - df['not_distracted'].min()) / h)
print(b)
sns.displot(data = df['not_distracted'], bins = b)
```
